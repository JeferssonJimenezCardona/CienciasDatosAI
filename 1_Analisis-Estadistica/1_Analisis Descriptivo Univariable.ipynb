Digamos que nos gustaría empalmar nuestro marco de datos y seleccionar solo porciones específicas de nuestros datos. Hay tres maneras diferentes de hacerlo.

1. .loc()
2. .iloc()
3. .ix()

## .loc() 
.loc() toma dos operadores individuales/lista/rango separados por ','. El primero indica la fila y el segundo indica las columnas.

```python
da.loc[:,"WTINT2YR"]
```

```python
da.loc[10:15]
```

.iloc() 
* .iloc() es un corte basado en enteros, mientras que .loc() usaba etiquetas/nombres de columna. Aquí hay unos ejemplos

```python
da.iloc[:4]
```

```python
da.iloc[1:5, 2:4]
```

```python
da.WTINT2YR.unique() # Listar valores únicos en la columna
```

```python
da.loc[:,["WTINT2YR", "BMXARML"]] # Filtrar una columnas en especifico
```

```python
pivot=da.groupby(["WTINT2YR", "BMXARML"]).size()  # Tabla Dinamica
pivot
```
